```python
import pandas as pd
pd.options.display.max_columns = 100
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sb
import csv
from matplotlib.ticker import StrMethodFormatter
import seaborn as sb
import numpy as np
import pandas as pd
```

```python
mozilla_df = pd.read_csv('mozilla_data.csv',header=0,encoding = 'unicode_escape')
mozilla_df.head()
```

#### The first visulalization that I created shows the languages that the participants of the survey use. The reason I chose to use a line graph to show the different languages is because I thought that this did a better job of showing the difference in languages than a bar chart would. Seeing the difference in the slope looks a lot different than how it would for a bar chart. I think the line graph does a better job of showing the differnce between all of the languages, but the bar chart is better for comparing one language to another. I wanted to do a line graph to try and show the how all of the languages that the participants use compare to each other. 

```python
#Line graph that shows the which language participants speak
mozilla_chart1 = mozilla_df['Language'].value_counts()
mozilla_chart1.plot(kind = 'line')
plt.title('Participant languages')
plt.ylabel('Number of participants')
plt.xlabel('Languages')
```

#### This bar chart shows who participants think is most responsible for teaching them internet safety. I used a horizaontal bar chart becasue of how long the labels were. On a normal bar chart they all would overalp one another. Same goes for the rest of the visualizations that I created. I didn't think that what I discovered from this visualization was very surprising.
```python
# Horizontal bar chart that shows who participants trust the most to teach them internet safety
mozilla_chart2 = mozilla_df['Who do you most trust to help you learn how to protect your safety, security and privacy online?'].value_counts()
mozilla_chart2.plot(kind = 'barh')
plt.title('Who participants trust the most for teaching them internet safety')
plt.xlabel('Number of participants')
```
